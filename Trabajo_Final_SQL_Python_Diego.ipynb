## 4. Visualizaciones

En esta sección se presentan algunas visualizaciones que permiten comprender mejor la distribución de los datos y las características de los solicitantes de crédito.

```python
fig = px.bar(
    df["Education"].value_counts().reset_index(),
    x="Education",
    y="count",
    title="Cantidad de solicitantes por nivel educativo",
    labels={"count": "Cantidad"}
)

fig.show()
```

**Interpretación:**  
Se observa que la mayoría de los solicitantes de préstamo cuentan con educación de nivel *Graduate*, mientras que un porcentaje menor corresponde a personas con nivel *Not Graduate*.

```python
fig = px.pie(
    df,
    names="Loan_Status",
    title="Estado de aprobación de préstamos",
    hole=0.5
)

fig.show()
```

**Interpretación:**  
La mayor parte de las solicitudes de préstamo fueron aprobadas (Loan_Status = Y), lo que representa aproximadamente el 69% del total de registros.

```python
fig = px.histogram(
    df,
    x="ApplicantIncome",
    nbins=30,
    title="Distribución de ingresos del solicitante"
)

fig.show()
```

**Interpretación:**  
La distribución de los ingresos presenta una concentración en los valores bajos y algunos casos con ingresos significativamente altos, lo que evidencia la presencia de valores atípicos.

```python
fig = px.box(
    df,
    y="LoanAmount",
    title="Distribución del monto del préstamo"
)

fig.show()
```

**Interpretación:**  
La mayoría de los montos de préstamo se concentran alrededor de un rango similar; sin embargo, existen algunos préstamos con montos considerablemente mayores, identificados como valores atípicos.

```python
fig = px.scatter(
    df,
    x="ApplicantIncome",
    y="LoanAmount",
    color="Loan_Status",
    title="Relación entre ingreso y monto del préstamo"
)

fig.show()
```

**Interpretación:**  
Se observa una tendencia donde los solicitantes con mayores ingresos pueden acceder a préstamos de mayor monto, aunque la relación no es completamente lineal y presenta una alta dispersión.

## 5. Consultas SQL con DuckDB

En esta sección se realizan consultas SQL sobre el DataFrame utilizando DuckDB para obtener información relevante del conjunto de datos.

### Consulta 1. Visualización de los primeros registros

```python
duckdb.sql("""
SELECT *
FROM df
LIMIT 10;
""").df()
```

### Consulta 2. Solicitantes con ingresos mayores a 10000

```python
duckdb.sql("""
SELECT Loan_ID,
       ApplicantIncome,
       LoanAmount,
       Loan_Status
FROM df
WHERE ApplicantIncome > 10000;
""").df()
```

### Consulta 3. Cantidad de solicitantes por nivel educativo

```python
duckdb.sql("""
SELECT Loan_ID,
       ApplicantIncome,
       LoanAmount,
       Loan_Status
FROM df
WHERE ApplicantIncome > 10000;
""").df()
```

### Consulta 4. Solicitantes con mayores ingresos

```python
duckdb.sql("""
SELECT Education,
       COUNT(*) AS Total_Solicitantes
FROM df
GROUP BY Education;
""").df()
```

### Consulta 5. Promedio de ingresos y préstamos según el estado de aprobación

```python
duckdb.sql("""
SELECT Loan_ID,
       ApplicantIncome,
       LoanAmount
FROM df
ORDER BY ApplicantIncome DESC
LIMIT 10;
""").df()
```

```python
duckdb.sql("""
SELECT Loan_Status,
       ROUND(AVG(ApplicantIncome),2) AS Promedio_Ingreso,
       ROUND(AVG(LoanAmount),2) AS Promedio_Prestamo
FROM df
GROUP BY Loan_Status;
""").df()
```
